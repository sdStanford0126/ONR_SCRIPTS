The goal of this script is to read from a slice the mesh levels and generate a line plot that reflects the meshsize at the given line (requires that link is present on the plot)

```python
from PIL import Image
import numpy as np
```

```python
def readImg(img_fname):
    img = Image.open(img_fname)
    np_img = np.asarray(img)
    return np_img

def buildpos(np_img,W,x_str,x_end):
    #doing for xz plane
    [Nz,Nx]=np_img.shape
    L = x_end - x_str
    dx = L/Nx
    vert_str = -W/2.0
    vert_end = W/2.0            
    x_vec = np.linspace(x_str,x_end,Nx,endpoint=True)
    vert_vec = np.linspace(vert_str,vert_end,Nz,endpoint=True)
    return x_vec, vert_vec,dx

def getGrayColorUnique(np_img):
    [npx,npy] = np_img.shape
    color_tuple=[]
    for i in range(npx):
        for j in range(npy):
            color_tuple.append(tuple(np_img[i,j,:]))
    unique_c_list = list(set(color_tuple))
    g_clist = []
    for c_tuple in unique_c_list:
        c_cur = np.asarray(c_tuple)
        if c_cur *3 == np.sum(c_cur):
            g_clist.append(tuple(c_cur))
    np.savetxt("gray_scale_unique_color.txt",g_clist)
    return g_clist
```
